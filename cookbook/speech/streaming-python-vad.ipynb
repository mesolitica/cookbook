```python
import requests

with open('test.mp3', 'wb') as fopen:
    r = requests.get('https://huggingface.co/datasets/mesolitica/speech-test-set/resolve/main/test.mp3')
    fopen.write(r._content)
```

```python
your_api_key = ''
```

```python
import aiohttp
original_headers = {
    'Authorization': f'Bearer {your_api_key}',
}

audio_file = open('test.mp3', 'rb')
data_form = {
    'file': audio_file,
    'model': 'base',
    'language': 'ms',
    'stream': '1',
    'chunking_method': 'vad',
    'vad_method': 'silero',
    'minimum_silent_ms': '150',
    'minimum_trigger_vad_ms': '1000',
}
async with aiohttp.ClientSession() as session:
    async with session.post(
        'https://api.mesolitica.com/audio/transcriptions',
        headers=original_headers,
        data=data_form,
    ) as r:
        if r.status == 200:
            async for line in r.content:
                line = line.decode().strip()
                print(line)
```
